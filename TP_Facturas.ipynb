# [dbo].[sp_ActualizarPrecios]

```sql
/****** Object:  StoredProcedure [dbo].[sp_ActualizarPrecios]    Script Date: 11/6/2024 18:22:04 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[sp_ActualizarPrecios]
	@pPorcentaje DECIMAL
AS
BEGIN
UPDATE [dbo].[Productos]
   SET precio = precio * (1 + @pPorcentaje / 100)
END
GO
```

# [dbo].[sp_BorrarTelefono]

```sql
/****** Object:  StoredProcedure [dbo].[sp_BorrarTelefono]    Script Date: 11/6/2024 18:22:04 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[sp_BorrarTelefono]
	@pidTelefono int

AS
BEGIN
	DECLARE @dni varchar(10)
	DECLARE @numeroTelefonos int

	SELECT @dni				= dni FROM Telefonos WHERE idTelefono = @pidTelefono
	SELECT @numeroTelefonos = COUNT(*)  FROM Telefonos WHERE dni = @dni

	IF @numeroTelefonos > 1
	BEGIN
		DELETE FROM [dbo].[Telefonos]
		WHERE idTelefono = @pidTelefono
	END
	ELSE
	PRINT 'No se puede borrar este telefono'
END
GO
```

# [dbo].[sp_InsertarDetalle]

```sql
/****** Object:  StoredProcedure [dbo].[sp_InsertarDetalle]    Script Date: 11/6/2024 18:22:04 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[sp_InsertarDetalle]
	@pIdVenta int,
	@pIdProducto int,
	@pCantidad int
AS
BEGIN

DECLARE @stock int
SELECT @stock= stock FROM Productos WHERE idProducto = @pIdProducto
IF @stock > 5
	IF @stock > @pCantidad
		BEGIN
		INSERT INTO [dbo].[DetalleVentas]
				   ([idVenta]
				   ,[idProducto]
				   ,[cantidad])
			 VALUES
				   (@pIdVenta,
					@pIdProducto,
					@pCantidad)

			UPDATE Productos
            SET stock = stock - @pCantidad
            WHERE idProducto = @pIdProducto
		END
	ELSE
		PRINT 'No se puede realizar la venta'
		
ELSE
	PRINT 'Falta de stock'
END
GO
```

# [dbo].[sp_ListarVentasXCliente]

```sql
/****** Object:  StoredProcedure [dbo].[sp_ListarVentasXCliente]    Script Date: 11/6/2024 18:22:04 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[sp_ListarVentasXCliente]
	@pDniCliente varchar(10),
	@pFecha1 date,
	@pFecha2 date
AS
BEGIN

SELECT [idVenta]
      ,[fecha]
      ,[dniCliente]
  FROM Ventas
  WHERE dniCliente= @pDniCliente
  AND Ventas.fecha BETWEEN @pFecha1 AND @pFecha2

END
GO
```

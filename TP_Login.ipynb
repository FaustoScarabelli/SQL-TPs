# [dbo].[sp_EliminarPerfil]

# [dbo].[sp_EliminarUsuario]

```sql
/****** Object:  StoredProcedure [dbo].[sp_EliminarUsuario]    Script Date: 11/6/2024 18:20:30 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[sp_EliminarUsuario]
	@pNombreUsuario varchar(50)
AS
BEGIN

DELETE FROM [dbo].[Usuarios]
      WHERE nombreUsuario = @pNombreUsuario
END
GO
```

# [dbo].[sp_ListarUsuariosxPerfil]

```sql
/****** Object:  StoredProcedure [dbo].[sp_ListarUsuariosxPerfil]    Script Date: 11/6/2024 18:20:30 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[sp_ListarUsuariosxPerfil]
	@pTipoUsuario varchar(50)
AS
BEGIN
	SELECT * FROM Usuarios u INNER JOIN TipoUsuario t ON u.tipoUsuarioId = t.tipoUsuarioId
	WHERE t.tipoUsuario = @pTipoUsuario
END
GO
```

# [dbo].[sp_Login]

```sql
/****** Object:  StoredProcedure [dbo].[sp_Login]    Script Date: 11/6/2024 18:20:30 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[sp_Login]
		@pNombreUsuario varchar(30) = null,
		@pContraseña varchar(30) = null

AS
BEGIN
	IF EXISTS (SELECT 1 FROM Usuarios WHERE nombreUsuario = @pNombreUsuario AND contraseña = @pContraseña)
	BEGIN
		SELECT * FROM Usuarios WHERE nombreUsuario = @pNombreUsuario AND contraseña = @pContraseña
	END
	ELSE
	BEGIN
		PRINT 'Usuario o contraseña incorrecta'
	END
END

GO
```

# [dbo].[sp_ModificarUsuario]

```sql
/****** Object:  StoredProcedure [dbo].[sp_ModificarUsuario]    Script Date: 11/6/2024 18:20:30 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[sp_ModificarUsuario]
	@pNombreUsuario varchar(59),
	@pContraseña varchar(30) = NULL,
	@pApellido varchar(30) = NULL,
	@pNombre varchar(30) = NULL,
	@pMail varchar(50) = NULL,
	@pTipoUsuarioId int = NULL
AS
BEGIN

UPDATE [dbo].[Usuarios]
   SET [contraseña] = IIF(@pContraseña is null, contraseña, @pContraseña)
      ,[apellido] = IIF(@pApellido is null, apellido, @pApellido)
      ,[nombre] = IIF(@pNombre is null, nombre, @pNombre)
      ,[mail] = IIF(@pMail is null, mail, @pMail)
      ,[tipoUsuarioId] = IIF(@pTipoUsuarioId is null, tipoUsuarioId, @pTipoUsuarioId)
 WHERE nombreUsuario = @pNombreUsuario
END
GO
```

# [dbo].[sp_RegistrarUsuario]

```sql
/****** Object:  StoredProcedure [dbo].[sp_RegistrarUsuario]    Script Date: 11/6/2024 18:20:30 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

Create Procedure [dbo].[sp_RegistrarUsuario]
	@pUsuarioId varchar(30),
	@pContraseña varchar(30),
	@pApellido varchar(30),
	@pNombre varchar(30),
	@pMail varchar(50),
	@pTipoUsuarioId int

AS

INSERT INTO [dbo].[Usuarios]
           ([nombreUsuario]
           ,[contraseña]
           ,[apellido]
           ,[nombre]
           ,[mail]
           ,[tipoUsuarioId])
     VALUES
			(
			 @pUsuarioId,
			 @pContraseña,
			 @pApellido ,
			 @pNombre,
			 @pMail,
			 @pTipoUsuarioId
			 )
    

GO
```
